```python
#affichage de l'image

from PIL import Image
import numpy as np
imgpil = Image.open("distance_to_object_test.jpg")
print(imgpil.size)

#sous forme de tableau
img = np.array(imgpil)
img

import matplotlib.pyplot as plt
plt.imshow(img)
plt.show()

#calcul de la distance en prenant des données du type [xmin,xmax,ymin,ymax]

#f = focale de l'appareil
#img_h = taille de l'image en pixels (longueur ou largeur suivant la situation)
#x_min x_max = taille de l'objet en pixels
#real_h =  hauteur de l'objet dans le vrai monde (en mm)
#sens_h = taille du capteur 

def distance_to_object(f,img_h,x_min,x_max,real_h,sens_h):
    return f*real_h*img_h/(sens_h*abs(x_min-x_max))
```

```python
#calculs expérimentaux Iphone 11
sensor_height=6.5*15.86*10
real_distance=8*15,86*10
real_height=15.86*10
focal = 4.5 #test avec teleobjectif
#sur la photo en pixels
a=int(imgpil.size[0])
b=int(imgpil.size[1])

image_height=250 #en regardant à peu près le graphe plt.imshow(img)

distance_to_object(focal,b,0,image_height,real_height,sensor_height)

#données pas dingues, à réessayer avec un appareil photo
```
